```python
from ultralytics import YOLO

# Load YOLOv8 model (pretrained on COCO)
model = YOLO("yolov8n.pt")

# Train model on number plate dataset
model.train(data="/content/drive/MyDrive/Car Plate/data.yaml", epochs=40, imgsz=640)
```

```python
model.val()
```

```python
from google.colab.patches import cv2_imshow
import cv2

# Run inference
results = model("/content/cartest.jpeg")

# Get the best bounding box (highest confidence)
best_box = None
best_conf = 0

for box in results[0].boxes:
    conf = box.conf.item()  # Confidence score
    if conf > best_conf:
        best_conf = conf
        best_box = box.xyxy[0]  # Bounding box coordinates (x_min, y_min, x_max, y_max)

# Load image
image = cv2.imread("/content/cartest.jpeg")

# Draw the best bounding box
if best_box is not None:
    x_min, y_min, x_max, y_max = map(int, best_box)  # Convert to integers
    cv2.rectangle(image, (x_min, y_min), (x_max, y_max), (0, 255, 0), 3)  # Draw green box

# Display the image in Colab
cv2_imshow(image)
```

```python
import cv2
from google.colab.patches import cv2_imshow
import easyocr

# Load image
image_path = "/content/cartest.jpeg"
image = cv2.imread(image_path)

# Initialize EasyOCR Reader
reader = easyocr.Reader(['en'])
results = reader.readtext(image)

#Draw bounding boxes on the detected text
for (bbox, text, prob) in results:
    top_left, bottom_right = tuple(map(int, bbox[0])), tuple(map(int, bbox[2]))
    cv2.rectangle(image, top_left, bottom_right, (0, 255, 0), 2)
    cv2.putText(image, text, (top_left[0], top_left[1] - 10),
                cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 255, 0), 2)
cv2_imshow(image)
```
